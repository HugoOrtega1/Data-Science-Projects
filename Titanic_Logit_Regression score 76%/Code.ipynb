## The Challenge <a id="cell1"></a>
The sinking of the Titanic is one of the most infamous shipwrecks in history.

On April 15, 1912, during her maiden voyage, the widely considered “unsinkable” RMS Titanic sank after colliding with an iceberg. Unfortunately, there weren’t enough lifeboats for everyone onboard, resulting in the death of 1502 out of 2224 passengers and crew.

While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this challenge, we ask you to build a predictive model that answers the question: “what sorts of people were more likely to survive?” using passenger data (ie name, age, gender, socio-economic class, etc).

```python
raw_data = pd.read_csv('train.csv')
raw_data
```

## Preprocessing Dataset<a id="cell5"></a>

### Dealing with missing values


```python
raw_data.isnull().sum()
```

```python
#Drop Cabin becasue it has vast amount of missing values as well as name and passenger ID becaue they will not fit my logit model well
raw_data = raw_data.drop(['Cabin','PassengerId','Name'], axis=1)
```

```python
# Replace na values with median of age column and rounding 
median = raw_data['Age'].median()
raw_data['Age'].fillna(median, inplace=True)
```

```python
raw_data.round()
```

```python
data_no_mv = raw_data.dropna(axis=0)
```

```python
data_no_mv.isnull().sum()
```

### Exploring the descriptive statistics of the variables

```python
data_no_mv.describe(include='all')
```

### Creating binary columns


```python
##using pandas cut and bins to create four different age groups and checking for missing values
bins= [0,14,24,65,99]
labels = ['Child','Teen','Young Adult', 'Senor']
data_no_mv['AgeGroup'] = pd.cut(data_no_mv['Age'], bins=bins, labels=labels, right=False)
data_no_mv1=data_no_mv.drop(['Age'], axis=1)

```

```python
data_no_mv1.isnull().sum()
```

```python
data_no_mv1['Child']=data_no_mv1['AgeGroup'].map({'Child':1,'Teen':0,'Young Adult':0,'Senor':0})
data_no_mv1['Teen']=data_no_mv1['AgeGroup'].map({'Child':0,'Teen':1,'Young Adult':0,'Senor':0})
data_no_mv1['Young Adult']=data_no_mv1['AgeGroup'].map({'Child':0,'Teen':0,'Young Adult':1,'Senor':0})
data_no_mv1['Senor']=data_no_mv1['AgeGroup'].map({'Child':0,'Teen':0,'Young Adult':0,'Senor':1})
data_no_mv2=data_no_mv1.drop(['AgeGroup'], axis=1)
```

```python
# Creating three new ticket classes  as well as mapping male and females to binary values 
data_no_mv2['PclassL']=data_no_mv2['Pclass'].map({1:'1',2:'0',3:'0'})
data_no_mv2['PclassM']=data_no_mv2['Pclass'].map({1:'0',2:'1',3:'0'})
data_no_mv2['PclassU']=data_no_mv2['Pclass'].map({1:'0',2:'0',3:'1'})
data_no_mv3 = data_no_mv2.drop(['Pclass','Fare','Ticket'], axis=1)

data_no_mv3['Sex'] = data_no_mv3['Sex'].map({'male':1,'female':0})

```

```python
data_no_mv3
```

### Creating dummies for categorical variables 

```python
data_with_dummies = pd.get_dummies(data_no_mv3, drop_first=True)
```

```python
data_with_dummies.head()
```

## Logit Regression Model <a id="cell6"></a>

### Declaring input and depdent variables

```python
targets = data_with_dummies['Survived']
inputs = data_with_dummies.drop(['Survived'], axis=1)
```

 ### Train_Test_Split data 
Setting a 80/20 split, splitting the training data into 80 and the test data to 20 with a random state of 9

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,targets,test_size= 0.20,random_state=9)
```

### Fitting Logit Model
I use a logit Model to train the data

```python
reglog =sm.Logit(y_train,x_train)
logitreg =reglog.fit()
```

### Logit Regression Results


```python
logitreg.summary()
```

## Accuracy Score of model <a id="cell7"></a>

```python
sc = np.array(logitreg.pred_table())
accuracy_score = (sc[0,0]+sc[1,1])/sc.sum()
accuracy_score
```
